```python
import pandas as pd
df = pd.read_csv(r'C:\Users\ADMIN\Desktop\Databyte_task1.csv')
df.head()
```

## Show the Total profit of all months using a line plot 

```python
import matplotlib.pyplot as plt
plt.plot(df['month_number'],df['total_profit'])
plt.show()
```

## Show all product sales data using a multiline plot 

```python
plt.plot(df['month_number'],df['facecream'])
plt.plot(df['month_number'],df['facewash'])
plt.plot(df['month_number'],df['toothpaste'])
plt.plot(df['month_number'],df['bathingsoap'])
plt.plot(df['month_number'],df['shampoo'])
plt.plot(df['month_number'],df['moisturizer'])
plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.show()
```

## Calculate total sales data for last year for each product and show it using a Pie chart

```python
df.sum(axis = 0, skipna = True)['facecream']
```

```python
slices = [df.sum(axis = 0, skipna = True)['facecream'],df.sum(axis = 0, skipna = True)['facewash'],df.sum(axis = 0, skipna = True)['toothpaste'],df.sum(axis = 0, skipna = True)['bathingsoap'],df.sum(axis = 0, skipna = True)['shampoo'],df.sum(axis = 0, skipna = True)['moisturizer']]
names = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
plt.pie(slices, labels = names)
plt.show()
```
