```python
import csv

with open('CountryCode.csv', mode='r') as infile:
    reader = csv.reader(infile)
    with open('CountryCode_new.csv', mode='w') as outfile:
        writer = csv.writer(outfile)
        mydict = {rows[2]:rows[1] for rows in reader}
```

```python
mydict
```

```python
def idx(string):
    i = 0
    for key in mydict:
        if(key == string):
            return i
        else:
            i += 1
```

```python
CC1 = input()
CC2 = input()
for key in mydict:
    if(idx(key)>idx(CC1) and idx(key)<idx(CC2)):
        print(mydict.get(key))
```
